```python
import os

num_cell_list = [10,20,40]

for num_cell in num_cell_list:
    path = os.path.join("cases","case_"+str(num_cell))
    os.system("cp -r base_case/ "+path)
    # goto blockMesh file, read in, replace text 
    block_mesh_path = os.path.join(path,"system","blockMeshDict")
    with open(block_mesh_path,"r") as file:
        filedata = file.read()
        # replace text
        filedata = filedata.replace("_res_x",str(num_cell)).replace("_res_y",str(num_cell))
        # write file 
    with open(block_mesh_path,"w") as file:
        file.write(filedata)
    file.close()
    os.system("./"+path+"/Allrun")
```

```python
import os 
import pandas as pd

path_cases = "cases"

case_list = os.listdir(path_cases)

df_collect = pd.DataFrame()

count = 0 

for case in case_list:
    case_post_path = os.path.join(path_cases,case,"postProcessing")
    post_object_list = os.listdir(case_post_path)
    for post_object in post_object_list:
        post_object_path = os.path.join(case_post_path,post_object)
        time_list = os.listdir(post_object_path)
        for time in time_list:
            line_object_path = os.path.join(post_object_path,time)
            data_object_list = os.listdir(line_object_path)
            for data_object in data_object_list:
                data_object_path = os.path.join(line_object_path,data_object)
                #print(data_object_path)
                df_data = pd.read_csv(data_object_path)
                data_object_name = data_object.split("_")[0]+data_object.split("_")[1]
                #print(data_object_name)
                index = pd.MultiIndex.from_product([[case],[post_object],[time],[data_object_name],df_data.columns.to_list()])
                columns = df_data.index
                df_sub = pd.DataFrame(index=index,columns=columns,data=df_data.transpose().values)
                df_collect = df_collect._append(df_sub)
                
# df_collect.to_excel("data.xlsx")
df_collect.to_csv("data.xlsx")
# df_collect.to_hdf("data.h5",'df')
```

```python
# Alternative 1 using glob
import os 
import glob

root_dir = os.getcwd().replace("\\","/")+"/"

csv_files = glob.glob(root_dir+"cases/*/postProcessing/*/*/*.csv")

df_collect = pd.DataFrame()

for csv_file_name in csv_files:
    case = csv_file_name.split("/")[10]
    post_object_name = csv_file_name.split("/")[12]
    time = csv_file_name.split("/")[13]
    filename = csv_file_name.split("/")[14]
    data_object_name = filename.split("_")[0]+filename.split("_")[1]
    df_data = pd.read_csv(csv_file_name)
    index = pd.MultiIndex.from_product([[case],[post_object],[time],[data_object_name],df_data.columns.to_list()])
    columns = df_data.index
    df_sub = pd.DataFrame(index=index,columns=columns,data=df_data.transpose().values)
    df_collect = df_collect.append(df_sub)
    
df_collect.to_excel("data.xlsx")

df_collect.to_hdf("data.h5",'df')
```
