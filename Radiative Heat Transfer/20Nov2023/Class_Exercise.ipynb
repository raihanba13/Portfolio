```python
import os 
import glob 
import re

root_dir = os.getcwd().replace("\\", "/") + "/"
path_cases = "cases/"
case_list = os.listdir(path_cases)

df_collect = pd.DataFrame()

case_list

csv_files = glob.glob(root_dir + 'cases/case_*/postProcessing/lines/*/*.csv')

df_collect = pd.DataFrame()

for csv_file in csv_files:
    rel_path = csv_file.replace(root_dir, "").replace("\\", "/")
    pattern = r"cases\/(.*?)\/.*?\/(.*?)\/(.*?)\/(.*?)\_(.*?)\_.*?\.csv"
    matches = re.finditer(pattern, rel_path, re.MULTILINE)
    
    for match in matches:
        case = match[1]
        post_object = match[2]
        time = match[3]
        data_object_name = match[4] + match[5]
        # print(case, post_object, time, data_object_name )
        
    df_data = pd.read_csv(csv_file)
    index = pd.MultiIndex.from_product([[case],[post_object],[time],[data_object_name],df_data.columns.to_list()])
    columns = df_data.index
    df_sub = pd.DataFrame(index=index,columns=columns,data=df_data.transpose().values)
    # append is changed to _append in newer pandas and concat should be used
    df_collect = df_collect._append(df_sub)


df_collect.to_csv("data2.csv")

```
