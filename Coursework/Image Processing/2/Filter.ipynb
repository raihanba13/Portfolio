##### We will discuss different type of filters, there usage.

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

Low pass filters can help in removing noise, blurring image and High pass filters help in finding edges in image

```python
# Linear filtering method, filter2D

kernel_list = [3, 11, 21]
img = cv.imread('test_1.JPG')

plt.figure()
plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])


for i in kernel_list:
    kernel = np.ones((i,i),np.float32)/i**2
    dst = cv.filter2D(img,-1,kernel)
    
    plt.figure()
    plt.imshow(dst),plt.title(f'Averaging, kernel size={i}')
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

```python
# Linear filtering method, Gaussian Blur
# Advantage:
# -> very effective to remove Gaussian noise
# -> The weights give higher significance to pixels near the edge (reduces edge blurring)
# -> Low coumputational cost and Rotationally symmetric

kernel_list = [3, 11, 21]
img = cv.imread('test_1.JPG')

plt.figure()
plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])


for i in kernel_list:
    dst = cv.GaussianBlur(img, (i,i),0)
    
    plt.figure()
    plt.imshow(dst),plt.title(f'Gaussian, kernel size={i}')
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

```python
# Nonlinear, Bilateral Filter
# Advantage:
# -> It removes most texture, noise, and fine details 
# -> They are non iterative

kernel_list = [30, 110, 840]
img = cv.imread('test_1.JPG')

plt.figure()
plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

"""
    From doc,

    Sigma values: For simplicity, you can set the 2 sigma values to be the same. If they are small (< 10), the filter will not have much effect, whereas if they are large (> 150), they will have a very strong effect, making the image look "cartoonish".

    Filter size: Large filters (d > 5) are very slow, so it is recommended to use d=5 for real-time applications, and perhaps d=9 for offline applications that need heavy noise filtering.
"""


for i in kernel_list:
    dst = cv.bilateralFilter(img, 5, i, i)
    
    plt.figure()
    plt.imshow(dst),plt.title(f'Bilateral Filter, filter size={i}')
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

High pass filter is a filter that passes only those signals whose frequencies are higher than cutoff frequencise
thereby attenuating signals of lower frequencies. This is useful to sharpen the details.
By applying these filters over an image, we can exaggerate every tiny part of details
in an image (i.e. finding the edges). 

```python
img = cv.imread('test_1.JPG')
laplacian_img = cv.Laplacian(img, -1, ksize=11, scale=1, delta=0, 
                          borderType=cv.BORDER_DEFAULT)

plt.figure()
plt.imshow(img),plt.title('Original')
plt.xticks([])
plt.yticks([])

plt.figure()
plt.imshow(laplacian_img),plt.title('Transformed using Laplacian')
plt.xticks([])
plt.yticks([])
```
