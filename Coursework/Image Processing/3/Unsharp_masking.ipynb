We will implement unsharp masking using the steps mentioned in lecture 4, slide 32

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline

img = cv.imread('test_1.JPG')
```

```python
# Step 1: Blur the original image

plt.figure()
plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

# ksize
ksize = (10, 10)
blur_img = cv.blur(img, ksize) 

plt.figure()
plt.imshow(blur_img),plt.title('Blur')
plt.xticks([]), plt.yticks([])
```

```python
# Step 2: Subtract the blurred image from the original
diff_img = img - blur_img

plt.figure()
plt.imshow(diff_img),plt.title('Difference of Original and blur image')
plt.xticks([]), plt.yticks([])

# diff_img is called mask
```

```python
# Step 3: Add the mask to the original
original_added_mask = img + diff_img

plt.figure()
plt.imshow(original_added_mask),plt.title('Addition of mask to the original')
plt.xticks([]), plt.yticks([])
```

```python
# Step 4: Now we try to add a weighted portion of the mask back to the original image
# nclip can be used to map the value that are out of the valid RGB rrange
ks = [1, 3, 2, -1, -2]
g_images = []

for k in ks:
    g_images.append(img + k*diff_img)
    
    if k == 1:
        title = "Unsharp masking"
    elif k > 1:
        title = "Highboost filtering"
    elif k < 1:
        title = "Reduced contribution of the unsharp masking"
    
    plt.figure()
    plt.imshow(g_images[-1]),plt.title(title+f", k={k}")
    plt.xticks([]), plt.yticks([])
    
```
